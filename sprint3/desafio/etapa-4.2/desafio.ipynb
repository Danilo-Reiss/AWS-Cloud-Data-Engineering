## Item 1
1) Leia o arquivo csv ```googleplaystore.csv``` e realize as seguintes atividades sobre o dataset **utilizando as bibliotecas Pandas e Matplotlib.** Remova as linhas duplicadas.

````python
import pandas as pd
import matplotlib.pyplot as plt

df_google = pd.read_csv('googleplaystore.csv')
display(df_google)
df_google.drop_duplicates(inplace=True)
print(df_google.shape)
````

## Tratamento 1
* Substituindo/eliminando valores nulos e convertendo tipos de dados.

````python
valores_nulos = df_google.isnull().any()
print(valores_nulos)

x = df_google['Rating'].mean()
df_google['Rating'].fillna(x, inplace=True)
df_google.dropna(inplace=True)

valores_nulos = df_google.isnull().any()
print(valores_nulos)
print(df_google.shape)

valores_distintos = df_google['Installs'].unique()
display(valores_distintos)

df_google['Installs'] = df_google['Installs'].str.slice(stop=-1)
df_google['Installs'] = df_google['Installs'].str.replace(',', '')
df_google['Installs'] = df_google['Installs'].astype(int)
display(df_google)
````

# Item 2
2) Faça um gráfico de barras contendo os top 5 apps por número de instalação.

````python
top5 = df_google[['App','Installs']].nlargest(5, 'Installs')
print(top5)

plt.figure(figsize=(20, 6))
plt.bar(top5['App'], top5['Installs'], color='skyblue', width=0.3)
plt.axis(ymin=0,ymax=1500000000)
plt.xlabel('Aplicativos')
plt.ylabel('Num. de instalações (em bilhões)')
plt.show()
````

# Item 3
3) Faça um gráfico de pizza (pie chart) mostrando as categorias de apps existentes no dataset de acordo com a frequência em que elas aparecem.

````python
freq_categ = df_google['Category'].value_counts(normalize=True) * 100
porcentagem = pd.Series(freq_categ)
display(porcentagem)

plt.figure(figsize=(20,20))
plt.pie(porcentagem, labels=porcentagem.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.show()
````

# Tratamento 2
* Convertendo tipos de dados.

````python
print(df_google['Price'].unique())

df_google['Price'] = df_google['Price'].str.replace('$','')
df_google['Price'] = df_google['Price'].astype(float)

print(df_google['Price'].unique())
````

````python
mais_caro = df_google[['App','Price']].nlargest(1,'Price')

display(mais_caro)
````

# Item 6
6) Mostre o top 10 apps por número de reviews bem como o respectivo número de reviews. Ordene a lista de forma decrescente por número de reviews.

````python
df_google['Reviews'] = df_google['Reviews'].astype(int)
lista_organizada = df_google.sort_values(by='Reviews', ascending=False)
lista_unique = lista_organizada.drop_duplicates(subset='App')
top10 = lista_unique[['App','Reviews']].nlargest(10,'Reviews')

display(top10)
````

### Gráfico 1 - Gráfico de linhas 

````python
plt.figure(figsize=(10, 6))
plt.plot(roblox_reviews.index, roblox_reviews['Reviews'], marker='o', color='skyblue', linestyle='--')
plt.title('Número de reviews do app ROBLOX a partir dos índices')
plt.xlabel('Índice')
plt.ylabel('Número de reviews (em milhão)')
plt.grid(True)
plt.show()
````

### Gráfico 2 - Gráfico de dispersão

````python
dado_generico = {'App':['App1','App2','App3'], 'Rating':[2.0, 1.4, 1.8]}
df = pd.DataFrame(dado_generico)
df_concatenado = pd.concat([menor_review, df])

plt.figure(figsize=(10, 6))
plt.scatter(df_concatenado['App'], df_concatenado['Rating'], color='red')
plt.title('Avaliação dos Aplicativos')
plt.xlabel('Aplicativos')
plt.ylabel('Avaliação')
plt.grid(True)
plt.show()
````
